```python
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=RuntimeWarning)
warnings.filterwarnings('ignore', category=UserWarning)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df_train = pd.read_csv("../input/tabular-playground-series-jun-2021/train.csv")
df_train.head()
```

```python
df_test = pd.read_csv("../input/tabular-playground-series-jun-2021/test.csv")
df_test.head()
```

```python
df_submission = pd.read_csv("../input/tabular-playground-series-jun-2021/sample_submission.csv")
df_submission.head()
```

```python
df_train.info()
df_test.info()
df_submission.info()
```

Problem type: MultiLabel Classification.
Steps: We have created our training and test data in separate dataframes. There are no null values. 

```python
df_train.drop(['id'],axis=1,inplace=True)
X=df_train.drop(['target'],axis=1,inplace=False).values
y=df_train['target'].values
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.25)


```

```python
clf= LogisticRegression()
clf.fit(X_train,y_train)
test_pred=clf.predict(X_val)
print(metrics.classification_report(y_val, test_pred))
  
```

```python
test=df_test.drop(['id'],axis=1).values
prediction=clf.predict_proba(test)
submission=pd.DataFrame(prediction,
                  columns=['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9'])
submission['id']=df_submission['id']
id= submission.pop('id')
submission.insert(0,'id',id)
```

```python
submission.to_csv("submission.csv",index=False)
submission.head()
```
